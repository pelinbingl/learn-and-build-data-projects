```python
import joblib
model = joblib.load("best_model_day6_tuned.pkl")
scaler = joblib.load("scaler_day6.pkl")
print("✅ Model yüklendi!")


df = pd.read_csv("cleaned_telco_churn.csv")

X = df.drop(columns=["Churn"])
y = df["Churn"]

X_scaled = scaler.transform(X)

print("✅ Model ve veri yüklendi.")
```

```python
# 2) PERFORMANS RAPORU
y_pred = model.predict(X_scaled)
y_proba = model.predict_proba(X_scaled)[:,1]

report = classification_report(y, y_pred, output_dict=True)
report_df = pd.DataFrame(report).T
roc_auc = roc_auc_score(y, y_proba)

print("📊 Genel Performans:")
display(report_df)
print(f"ROC-AUC: {roc_auc:.3f}")
```

```python
# 3) ROC ve Confusion Matrix
RocCurveDisplay.from_estimator(model, X_scaled, y)
plt.title("Final ROC Eğrisi – Telco Churn Modeli")
plt.show()

ConfusionMatrixDisplay.from_estimator(model, X_scaled, y, cmap="PuRd")
plt.title("Final Confusion Matrix – Telco Churn")
plt.show()
```

```python
# 4) ÖZELLİK ÖNEMLERİ
feat_imp = pd.DataFrame({
    "Feature": X.columns,
    "Importance": model.feature_importances_
}).sort_values(by="Importance", ascending=False)

plt.figure(figsize=(8,6))
sns.barplot(x="Importance", y="Feature", data=feat_imp.head(10))
plt.title("Final Model – En Önemli 10 Özellik")
plt.tight_layout()
plt.show()

feat_imp.head(10)
```

```python
# 5) RAPORU OLUŞTUR
with open("final_report_telco.md", "w", encoding="utf-8") as f:
    f.write("# 📊 Telco Customer Churn – Final Rapor\n\n")
    f.write("## Model Performansı\n")
    f.write(report_df.to_markdown())
    f.write(f"\n\n**ROC-AUC:** {roc_auc:.3f}\n")
    f.write("\n## En Önemli Özellikler\n")
    f.write(feat_imp.head(10).to_markdown())
    f.write("\n\n📁 Model: `best_model_day6_tuned.pkl`\n")
    f.write("\n💬 *Bu model, SMOTE ve GridSearchCV ile optimize edilmiş final sürümdür.*\n")
    f.write("\n🧠 Hazırlayan: Pelin Bingöl\n")

print("✅ Markdown raporu oluşturuldu: final_report_telco.md")
```
