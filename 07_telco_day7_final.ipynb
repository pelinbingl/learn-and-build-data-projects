```python
import joblib
model = joblib.load("best_model_day6_tuned.pkl")
scaler = joblib.load("scaler_day6.pkl")
print("‚úÖ Model y√ºklendi!")


df = pd.read_csv("cleaned_telco_churn.csv")

X = df.drop(columns=["Churn"])
y = df["Churn"]

X_scaled = scaler.transform(X)

print("‚úÖ Model ve veri y√ºklendi.")
```

```python
# 2) PERFORMANS RAPORU
y_pred = model.predict(X_scaled)
y_proba = model.predict_proba(X_scaled)[:,1]

report = classification_report(y, y_pred, output_dict=True)
report_df = pd.DataFrame(report).T
roc_auc = roc_auc_score(y, y_proba)

print("üìä Genel Performans:")
display(report_df)
print(f"ROC-AUC: {roc_auc:.3f}")
```

```python
# 3) ROC ve Confusion Matrix
RocCurveDisplay.from_estimator(model, X_scaled, y)
plt.title("Final ROC Eƒürisi ‚Äì Telco Churn Modeli")
plt.show()

ConfusionMatrixDisplay.from_estimator(model, X_scaled, y, cmap="PuRd")
plt.title("Final Confusion Matrix ‚Äì Telco Churn")
plt.show()
```

```python
# 4) √ñZELLƒ∞K √ñNEMLERƒ∞
feat_imp = pd.DataFrame({
    "Feature": X.columns,
    "Importance": model.feature_importances_
}).sort_values(by="Importance", ascending=False)

plt.figure(figsize=(8,6))
sns.barplot(x="Importance", y="Feature", data=feat_imp.head(10))
plt.title("Final Model ‚Äì En √ñnemli 10 √ñzellik")
plt.tight_layout()
plt.show()

feat_imp.head(10)
```

```python
# 5) RAPORU OLU≈ûTUR
with open("final_report_telco.md", "w", encoding="utf-8") as f:
    f.write("# üìä Telco Customer Churn ‚Äì Final Rapor\n\n")
    f.write("## Model Performansƒ±\n")
    f.write(report_df.to_markdown())
    f.write(f"\n\n**ROC-AUC:** {roc_auc:.3f}\n")
    f.write("\n## En √ñnemli √ñzellikler\n")
    f.write(feat_imp.head(10).to_markdown())
    f.write("\n\nüìÅ Model: `best_model_day6_tuned.pkl`\n")
    f.write("\nüí¨ *Bu model, SMOTE ve GridSearchCV ile optimize edilmi≈ü final s√ºr√ºmd√ºr.*\n")
    f.write("\nüß† Hazƒ±rlayan: Pelin Bing√∂l\n")

print("‚úÖ Markdown raporu olu≈üturuldu: final_report_telco.md")
```
