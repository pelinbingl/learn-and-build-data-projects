**Day 2 – Veri Temizleme ve Eksik Değer Analizi**

```python

# 🧹 1. KÜTÜPHANELERİN YÜKLENMESİ VE VERİNİN TEKRAR OKUNMASI

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use("ggplot")
sns.set_palette("pastel")

from google.colab import files
files.upload()

print("✅ Veri başarıyla yüklendi!")
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

print("Boyut:", df.shape)
df.head()
```

```python

#  2. EKSİK DEĞER ANALİZİ

# Eksik değerleri kontrol et
print("Eksik Değer Sayısı:")
print(df.isnull().sum())

# Bazı değerler boş değil ama boşluk karakteri içerebilir
# Özellikle 'TotalCharges' sütununda bu durum sık görülür
print("\nBoşluk karakterli hücre sayısı (TotalCharges):",
      df["TotalCharges"].apply(lambda x: str(x).isspace()).sum())
```

```python

# ⚙️ 3. EKSİK VE HATALI DEĞERLERİ DÜZELTME

# 'TotalCharges' sütunundaki boşluk karakterlerini NaN yapalım
df["TotalCharges"] = df["TotalCharges"].replace(" ", np.nan)

# Bu NaN değerleri 0 ile dolduruyoruz
df["TotalCharges"] = df["TotalCharges"].fillna(0)

# Veri tipini float'a dönüştür
df["TotalCharges"] = df["TotalCharges"].astype(float)

print("✅ TotalCharges sütunu başarıyla düzeltildi!")
df.info()
```

```python

# 🧱 4. GEREKSİZ SÜTUNLARIN KALDIRILMASI

# 'customerID' modeli etkilemeyen bir kimlik bilgisi → kaldırıyoruz
df = df.drop(columns=["customerID"])

print("✅ Gereksiz sütunlar kaldırıldı.")
print("Yeni veri boyutu:", df.shape)
```

```python

# 🔄 5. KATEGORİK DEĞİŞKENLERİN HAZIRLANMASI


# Hedef değişken: 'Churn' → Yes/No yerine 1/0 yapalım
df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})

# Kategorik sütunları bulalım
cat_cols = df.select_dtypes(include=["object"]).columns
print("Kategorik sütunlar:", cat_cols.tolist())

# One-hot encoding (kategorik verileri sayısallaştırma)
df_encoded = pd.get_dummies(df, drop_first=True)

print("✅ Kategorik veriler encode edildi.")
print("Yeni boyut:", df_encoded.shape)
```

```python

# 🧾 6. TEMİZLENMİŞ VERİNİN KONTROLÜ

print("İlk 5 satır (Temizlenmiş veri):")
df_encoded.head()
```

```python

# 💾 7. TEMİZ VERİYİ KAYDETME

df_encoded.to_csv("cleaned_telco_churn.csv", index=False)
print("✅ Temiz veri başarıyla kaydedildi: cleaned_telco_churn.csv")
```
