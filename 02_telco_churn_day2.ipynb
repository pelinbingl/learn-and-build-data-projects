```python

# üßπ 1. K√úT√úPHANELERƒ∞N Y√úKLENMESƒ∞ VE VERƒ∞Nƒ∞N TEKRAR OKUNMASI

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use("ggplot")
sns.set_palette("pastel")

from google.colab import files
files.upload()

print("‚úÖ Veri ba≈üarƒ±yla y√ºklendi!")
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")

print("Boyut:", df.shape)
df.head()
```

```python

#  2. EKSƒ∞K DEƒûER ANALƒ∞Zƒ∞

# Eksik deƒüerleri kontrol et
print("Eksik Deƒüer Sayƒ±sƒ±:")
print(df.isnull().sum())

# Bazƒ± deƒüerler bo≈ü deƒüil ama bo≈üluk karakteri i√ßerebilir
# √ñzellikle 'TotalCharges' s√ºtununda bu durum sƒ±k g√∂r√ºl√ºr
print("\nBo≈üluk karakterli h√ºcre sayƒ±sƒ± (TotalCharges):",
      df["TotalCharges"].apply(lambda x: str(x).isspace()).sum())
```

```python

# ‚öôÔ∏è 3. EKSƒ∞K VE HATALI DEƒûERLERƒ∞ D√úZELTME

# 'TotalCharges' s√ºtunundaki bo≈üluk karakterlerini NaN yapalƒ±m
df["TotalCharges"] = df["TotalCharges"].replace(" ", np.nan)

# Bu NaN deƒüerleri 0 ile dolduruyoruz
df["TotalCharges"] = df["TotalCharges"].fillna(0)

# Veri tipini float'a d√∂n√º≈üt√ºr
df["TotalCharges"] = df["TotalCharges"].astype(float)

print("‚úÖ TotalCharges s√ºtunu ba≈üarƒ±yla d√ºzeltildi!")
df.info()
```

```python

# üß± 4. GEREKSƒ∞Z S√úTUNLARIN KALDIRILMASI

# 'customerID' modeli etkilemeyen bir kimlik bilgisi ‚Üí kaldƒ±rƒ±yoruz
df = df.drop(columns=["customerID"])

print("‚úÖ Gereksiz s√ºtunlar kaldƒ±rƒ±ldƒ±.")
print("Yeni veri boyutu:", df.shape)
```

```python

# üîÑ 5. KATEGORƒ∞K DEƒûƒ∞≈ûKENLERƒ∞N HAZIRLANMASI


# Hedef deƒüi≈üken: 'Churn' ‚Üí Yes/No yerine 1/0 yapalƒ±m
df["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})

# Kategorik s√ºtunlarƒ± bulalƒ±m
cat_cols = df.select_dtypes(include=["object"]).columns
print("Kategorik s√ºtunlar:", cat_cols.tolist())

# One-hot encoding (kategorik verileri sayƒ±salla≈ütƒ±rma)
df_encoded = pd.get_dummies(df, drop_first=True)

print("‚úÖ Kategorik veriler encode edildi.")
print("Yeni boyut:", df_encoded.shape)
```

```python

# üßæ 6. TEMƒ∞ZLENMƒ∞≈û VERƒ∞Nƒ∞N KONTROL√ú

print("ƒ∞lk 5 satƒ±r (Temizlenmi≈ü veri):")
df_encoded.head()
```

```python

# üíæ 7. TEMƒ∞Z VERƒ∞Yƒ∞ KAYDETME

df_encoded.to_csv("cleaned_telco_churn.csv", index=False)
print("‚úÖ Temiz veri ba≈üarƒ±yla kaydedildi: cleaned_telco_churn.csv")
```
